```python
import pandas as pd
import json
from collections import Counter
import matplotlib.pyplot as plt

# Suponha que voc tenha carregado o DataFrame em 'df'
df = pd.read_json('union_all_jsons.json')

def calculate_probabilities(dataframe):
    probabilities = {}

    # Filtrar linhas onde 'WannaCry' est na lista de malware_families
    wanna_cry_df = dataframe[dataframe['malware_families'].apply(lambda x: 'WannaCry' in x)]

    # Contar o nmero total de indicadores em todas as amostras do WannaCry
    total_indicators = sum(wanna_cry_df['indicators'].apply(len))

    # Iterar sobre as linhas e calcular as probabilidades
    for _, row in wanna_cry_df.iterrows():
        for indicator in row['indicators']:
            indicator_name = indicator['indicator']
            if indicator_name not in probabilities:
                probabilities[indicator_name] = 0
            probabilities[indicator_name] += 1 / total_indicators

    return probabilities

probabilities = calculate_probabilities(df)

# Classificar as probabilidades em ordem decrescente
sorted_probabilities = sorted(probabilities.items(), key=lambda x: x[1], reverse=True)

# Imprimir as probabilidades
for indicator, probability in sorted_probabilities:
    print(f'Indicator: {indicator}, Probability: {probability}')
```
