```python
import pandas as pd

# df_combined = pd.DataFrame()

df_wanna_cry = pd.read_json('combined_subscribed_wanna_cry.json')
df_wanna_cry = df_wanna_cry['results']

# df_combined = pd.concat([df_combined, df_wanna_cry], ignore_index=True)

df_general = pd.read_json('combined_subscribed_general.json')
df_general = df_general['results']

df_combined = pd.concat([df_wanna_cry, df_general], ignore_index=True)

df_combined
```

```python
# pulses = pd.read_json('wanna_cry_results.json')
# pulses

# Converter o DataFrame combinado em um arquivo JSON
union_all = df_combined.to_json(orient='records')

# Salvar o arquivo JSON combinado
with open('union_all_jsons.json', 'w', encoding='utf-8') as json_file:
    json_file.write(union_all)
```
